# Derivative of Matrix

## import library

````python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import matplotlib.colors as colors
````

## load image 

````python
I = img.imread('img/a.jpg')  
````

## check the size of image

````python
n_row = I.shape[0]
n_col = I.shape[1]

print(I.shape)
````

## convert the input image into gray scale if it is color

````python
if I.shape[2] == 3:
    red = I[:,:,0]
    blue = I[:,:,1]
    green = I[:,:,2]
    
    I =  (red * 0.299 + green * 0.587 + blue * 0.114)


````

## normalize input image so that the range of image is $[0, 1]$

````python
I = (I - np.min(I)) / (np.max(I) - np.min(I))
````

## generate a matrix to compute the derivative in $x$-direction

````python

Dx = np.zeros((n_col,n_col))
y = 0;
Dx[0][0] = -1
for x in range(1, n_col):
    Dx[x][y] = 1
    Dx[x][y+1] = -1
    y+=1
        
Dx
````

## compute the derivative of $I$ with respect to $x$-direction

````python
Ix = np.dot(I, Dx)

Ix
````

## generate a matrix to compute the derivative in $y$-direction

````python

Dy = np.zeros((n_row,n_row))
y = 0;
for x in range(0, n_row-1):
    Dy[x][y] = -1
    Dy[x][y+1] = 1
    y+=1

Dy[n_row-1][n_row-1] = -1
Dy
````

## compute the derivative of $I$ with respect to $y$-direction

````python
Iy = np.dot(Dy, I)

Iy
````

```git commit -a -m "compute the derivative in y-direction"``` \
```git push origin master```

## compute $L_2^2$ of the gradient of $I$

````python
Ix2 = np.square(Ix)
Iy2 = np.square(Iy)
````

````python
norm_gradient = np.sqrt(Ix2+Iy2)
norm_gradient
````

## define functions for the visualization

````python
N = 100
X, Y = np.mgrid[-3:3:complex(0, N), -2:2:complex(0, N)]

Z1 = np.exp(-(X)**2 - (Y)**2)
Z2 = np.exp(-(X * 10)**2 - (Y * 10)**2)
Z = Z1 + 50 * Z2
````

````python
def plot_image(I):
    
    plt.figure(figsize=(10,10))
    plt.imshow(I, cmap = 'gray')
    plt.title('input image')
    plt.axis('off')
        
def plot_image_derivative_x(Ix):
    
    plt.figure(figsize=(10,10))
    plt.imshow(Ix, cmap = 'gray',  norm=colors.LogNorm(vmin=Z.min(), vmax=Z.max()))
    plt.title('image derivative of x')
    plt.axis('off')

    
def plot_image_derivative_y(Iy):
    
    plt.figure(figsize=(10,10))
    plt.imshow(Iy, cmap = 'gray',  norm=colors.LogNorm(vmin=Z.min(), vmax=Z.max()))
    plt.title('image derivative of y')
    plt.axis('off')

    
def plot_norm_gradient(norm_gradient):

    plt.figure(figsize=(10,10))
    plt.imshow(norm_gradient, cmap = 'gray',  norm=colors.LogNorm(vmin=Z.min(), vmax=Z.max()))
    plt.title('norm gradient')
    plt.axis('off')
````
